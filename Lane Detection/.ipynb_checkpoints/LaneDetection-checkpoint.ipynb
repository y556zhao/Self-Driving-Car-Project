### Import libraries and read in test image

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline

img = mpimg.imread('./image/test.jpg')
plt.imshow(img)
```

### Canny Edge

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

# Apply Blur
kernel_size = 5 
blur_gray = cv2.GaussianBlur(gray_img,(kernel_size, kernel_size),0)

# Define our parameters for Canny 
low_threshold = 50
high_threshold = 150
edges = cv2.Canny(blur_gray, low_threshold, high_threshold)

# Display the image
plt.imshow(edges, cmap='Greys_r')
```

### Region Mask

```python
# create a black image
mask = np.zeros_like(edges)   
ignore_mask_color = 255 

imshape = img.shape
vertices = np.array([[(0,imshape[0]),(450, 290), (490, 290), (imshape[1],imshape[0])]], dtype=np.int32)
cv2.fillPoly(mask, vertices, ignore_mask_color)

masked_edges = cv2.bitwise_and(edges, mask)
plt.imshow(masked_edges)
```

### Hough Lines

```python
rho = 2 
theta = np.pi/180
threshold = 15     
min_line_length = 57
max_line_gap = 20  
line_image = np.copy(img)*0 

left_lane=[]
right_lane=[]

lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]),
                            min_line_length, max_line_gap)
for line in lines:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),10)
        slope, intercept = np.polyfit((x1,x2),(y1,y2),1) 

        if slope < 0:
            right_lane.append(slope)
        else :
            left_lane.append(slope)
color_edges = np.dstack((edges, edges, edges)) 

# Draw the lines 
# Calculates the weighted sum of two arrays
lines_edges = cv2.addWeighted(color_edges, 0.8, line_image, 1, 0) 
plt.imshow(lines_edges)
```

### Optimize lines

```python
left_average  = np.average(left_lane)
right_average  = np.average(right_lane)

x_left=0.20*imshape[1]
x_right=0.85*imshape[1]
y=0.6*imshape[0]
y_from_bottom=0.4*imshape[0]

left_line=[x_left,imshape[0],(x_left*left_average+y_from_bottom)/left_average,y]
right_line=[x_right,imshape[0],(x_right*right_average-y_from_bottom)/right_average,y]

print(left_line)
print(right_line)
```
